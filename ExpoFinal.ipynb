```python
### LIBRERIAS
from selenium import webdriver #Después de haber instalado selenium
from selenium.webdriver.support.ui import WebDriverWait #Librería para esperar tiempo para dar click
from selenium.webdriver.support import expected_conditions as EC #Librerías para navegación y extraccion de información
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time #Librería para pausar los códigos y darle tiempo a las páginas a que terminen de cargar
import pandas as pd #Con esta librería se hará dataframe de la información recopilada para su futuro envío por correo
import re #libreria para buscar determinadas secuencias de caracteres en los datos
```

```python
### PRODUCTO A BUSCAR
producto = input("Ingrese el producto a buscar: ")
```

```python
### METRO
# Opciones de navegación
options =  webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')
driver = webdriver.Chrome(options=options)
time.sleep(1)

# Inicializamos el navegador
driver.get("https://www.metro.pe/")


# Clicks para cerrar opciones de cookies en la pagina
WebDriverWait(driver, 20)\
    .until(EC.element_to_be_clickable((By.XPATH,"//button[@class='align-right secondary slidedown-button']")))\
    .click()

WebDriverWait(driver, 15)\
    .until(EC.element_to_be_clickable((By.XPATH,"//button[@class='swal2-confirm swal2-styled']")))\
    .click()

#Click para ingresar producto a buscar
WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,"//input[@id='search-autocomplete-input']")))\
    .send_keys(producto)

#Click para buscar todos los productos
WebDriverWait(driver, 15)\
    .until(EC.element_to_be_clickable((By.XPATH,"//a[@class='see-all']")))\
    .click()

#Mover la barra vertical de desplazamiento hacia abajo para cargar todos los productos
driver.execute_script("window.scrollBy(0, document.body.scrollHeight);")

#le damos tiempo al a que carguen todos los productos
time.sleep(15)

# Definiendo nombres y precios

def precios_metro():
    tarifas = driver.find_elements(By.XPATH,"//div[@class='product-item__bottom']") #Asignando espacio donde se encuentra la información a recopilar
    precios = [] #Creando lista de precios
    nombres = [] #Creando listas de nombres 
    for tarifa in tarifas: #Recorriendo tarifas
        nombre = tarifa.find_element(By.XPATH,".//a[@class='product-item__name']").text #Extrayendo texto de los nombres
        precio = tarifa.find_element(By.XPATH,".//span[@class='product-prices__value product-prices__value--best-price']").text #Extrayendo texto de los precios
        nombres.append(nombre) #Agregando a listas
        precios.append(precio)
    df_metro=pd.DataFrame({"Nombre_producto":nombres,"Precios_en_soles":precios, "Supermercado":["metro"]*len(nombres)}) #Creando DataFrame con la informacion extraida
    return df_metro
precios_metro()
```

```python
metro = precios_metro()[['Precios_en_soles']]#extraemos la informacion de los precios
m = metro.to_numpy().tolist()#convertiendo un dataframe en una lista
z = len(precios_metro())#calculando la longitud de la lista
listame = []#creando una lista
for i in range(z): #recorremos la lista
    x = m[i]
    x = "".join(x)#convertimos los datos en una cadena
    x = re.sub("[Ss]/[.]?\W?","",x)#eliminamos los signos de moneda y otros carácteres
    x = float(x)#convertimos los datos en flotantes
    listas = listame.append(x)#agregamos los datos a la lista
print(listame)
```

```python
### TOTTUS
# Opciones de navegación
options =  webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')
driver = webdriver.Chrome(options=options)
time.sleep(1)

# Inicializamos el driver
driver.get("https://www.tottus.com.pe/")

 
# Click para cerrar opciones de cookies en la pagina
WebDriverWait(driver, 40)\
    .until(EC.element_to_be_clickable((By.XPATH,"//button[@class='align-right secondary slidedown-button']")))\
    .click()

#Click para ingresar producto a buscar
WebDriverWait(driver, 10)\
    .until(EC.element_to_be_clickable((By.XPATH,"//input[@class='jsx-3290144217 desktopSearchBar']")))\
    .send_keys(producto)

#Click para buscar todos los productos
WebDriverWait(driver, 10)\
    .until(EC.element_to_be_clickable((By.XPATH,"//button[@class='jsx-3290144217 searchIcon']")))\
    .click()

#Le damos 10 segundos para que cargue
time.sleep(10)

def precios_tottus():
    casillas = driver.find_elements(By.XPATH,"//li[@class='product']")
    precios = []
    nombres = []
    for casilla in casillas:
        nombre = casilla.find_element(By.XPATH,".//h2[@class='jsx-369422875 name title mini']").text
        precio = casilla.find_element(By.XPATH,".//div[@class='jsx-4019956066 ProductPrice']").text
        nombres.append(nombre)
        precios.append(precio)
        df_tottus=pd.DataFrame({"Nombre_producto":nombres,"Precios_en_soles":precios, "Supermercado":["tottus"]*len(nombres)})
    return df_tottus
precios_tottus()
```

```python
tottus = precios_tottus()[['Precios_en_soles']]#extraemos la informacion de los precios
m = tottus.to_numpy().tolist()#convertiendo un dataframe en una lista
z = len(precios_tottus())#calculando la longitud de la lista
listato = []#creando una lista
for i in range(z):#recorremos la lista
    x = m[i]
    x = "".join(x)#convertimos los datos en una cadena
    x = re.sub("[Ss]/[.]?\W?","",x)#eliminamos los signos de moneda y otros carácteres
    x = re.sub(".[Uu][Nn]?","",x)#eliminamos el signo de unidad
    x = float(x)#convertimos los datos en flotantes
    listas = listato.append(x)#agregamos los datos a la lista
print(listato)
```
