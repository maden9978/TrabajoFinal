```python
### LIBRERIAS
from selenium import webdriver #Después de haber instalado selenium
from selenium.webdriver.support.ui import WebDriverWait #Librería para esperar tiempo para dar click
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time #Librería para pausar los códigos y darle tiempo a las páginas a que terminen de cargar
import pandas as pd #Con esta librería se hará dataframe de la información recopilada para su futuro envío por correo
```

```python
### PRODUCTO A BUSCAR
producto = input("Ingrese el producto a buscar: ")
```

```python
### METRO
# Opciones de navegación
options =  webdriver.ChromeOptions()
options.add_argument('--start-maximized') #Pantalla completa
options.add_argument('--disable-extensions') #Desactivar extensiones
driver = webdriver.Chrome(options=options) #Definiendo al driver
time.sleep(1)

# Inicializamos el navegador
driver.get("https://www.metro.pe/")


# Clicks para cerrar opciones de cookies en la pagina
WebDriverWait(driver, 20)\
    .until(EC.element_to_be_clickable((By.XPATH,"//button[@class='align-right secondary slidedown-button']")))\
    .click()

WebDriverWait(driver, 15)\
    .until(EC.element_to_be_clickable((By.XPATH,"//button[@class='swal2-confirm swal2-styled']")))\
    .click()

#Click para ingresar producto a buscar
WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,"//input[@id='search-autocomplete-input']")))\
    .send_keys(producto)

#Click para buscar todos los productos
WebDriverWait(driver, 15)\
    .until(EC.element_to_be_clickable((By.XPATH,"//a[@class='see-all']")))\
    .click()

#Mover la barra vertical de desplazamiento hacia abajo para cargar todos los productos
driver.execute_script("window.scrollBy(0, document.body.scrollHeight);")

#le damos tiempo al a que carguen todos los productos
time.sleep(15)

# Se crea diccionario para productos y su precio, para ello definimos

def precios_metro(): 
    tarifas = driver.find_elements(By.XPATH,"//div[@class='product-item__bottom']") #Asignando espacio donde se encuentra la información a recopilar
    precios = [] #Creando lista de precios
    for tarifa in tarifas: #Recorriendo tarifas
        nombre = tarifa.find_element(By.XPATH,".//a[@class='product-item__name']").text #Extrayendo texto de los nombres
        precio = tarifa.find_element(By.XPATH,".//span[@class='product-prices__value product-prices__value--best-price']").text #Extrayendo texto de los precios
        dict_tarifa={nombre:{'precio':precio}} #Creando diccionario
        precios.append(dict_tarifa) #Agregando listas a diccionario
        print(dict_tarifa)

```

```python
METRO = precios_metro() #Guardar la información en una variable 
```

```python
import imghdr #Librería para enviar archivos por correo
import smtplib #Librería para enviar correos desde Python
import ssl #Esta librería brinda mayor seguridad para trabajar con los correos
from email.message import EmailMessage #Con esta librería definimos el código para el emisor y receptor

# Define email mensajero y receptor
email_sender = input("escriba su correo: ")
email_password = input("contraseña: ")
email_receiver = input("correo receptor: ")

# Configurar el asunto y cuerpo de texto
subject = 'Precios'
body = """En este correo se le envía en formato csv a usted los precios de los tres supermercados a analizar
.
.
.
"""
#Definiendo parámetros de envío 
em = EmailMessage()
em['From'] = email_sender
em['To'] = email_receiver
em['Subject'] = subject
em.set_content(body)

# Adjuntar archivo
wit open('precios.csv','rb') as file:  #Agregando archivo csv donde se comparan los precios"
    file_data = file.read() #Informacion del archivo
    file_tipo = imghdr.what(file.name) #Tipo de archivo
    filer_nombre = file.name #Nombre del archivo
em.add_attachment(file_data, filename=file_nombre, subtype = file_tipo, maintype = "application") #Parámetros del archivo adjunto para su futuro envío

# Añadiendo SSL para mayor seguridad de navegación
context = ssl.create_default_context()

# Codigo para logear y mandar el email
with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as smtp: #Se define parámetros para el envío
    smtp.login(email_sender, email_password)
    smtp.sendmail(email_sender, email_receiver, em.as_string())
```
